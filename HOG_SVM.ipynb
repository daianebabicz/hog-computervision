**Instalação de dependências**

```python
from google.colab import drive
drive.mount('/content/gdrive')

from skimage import feature
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
import os
import cv2
```

**Implementando a função que extrai os descritores HoG:**

```python
def get_hog_features(img):
    # Redimensione a imagem para 64x128 pixels (padrão do HoG Descriptor)
    img = cv2.resize(img, (64, 128))

    # Calcula os recursos HoG
    hog_features = feature.hog(img, orientations=9, pixels_per_cell=(8, 8),
                               cells_per_block=(2, 2), block_norm='L2-Hys')

    return hog_features
```

**Carregar as imagens e extrair os descritores HoG:**

```python
def load_images_from_folder(folder):
    images = []
    labels = []
    for label, subfolder in enumerate(os.listdir(folder)):
        for filename in os.listdir(os.path.join(folder, subfolder)):
            img = cv2.imread(os.path.join(folder, subfolder, filename), cv2.IMREAD_GRAYSCALE)
            if img is not None:
                hog_features = get_hog_features(img)
                images.append(hog_features)
                labels.append(label)
                if len(images) <= 5:  # apenas para os primeiros 5 vetores de características
                    print(hog_features)
    return images, labels

images, labels = load_images_from_folder('/content/gdrive/MyDrive/Visão Computacional/Trabalho HoG/fruits-dataset')
```

**Dividindo o conjunto de dados em treino e teste:**

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.3, random_state=400)
```

**Treinar o classificador SVM:**

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
```

**Testar o modelo e extrair as métricas:**

```python
y_pred = clf.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred, average='micro'))
```
